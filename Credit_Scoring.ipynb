```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.optimizers import Adam
import numpy as np
import pandas as pd
```

```python
creditset = pd.read_csv("creditset.csv")
features = creditset[["income","age","loan"]]
target = creditset.default
# print(features.shape)
features = np.array(features).reshape(-1,3)
# print(features.shape)
target = np.array(target).reshape(-1,1)

encoder = OneHotEncoder()
targets = encoder.fit_transform(target).toarray()

feature_train , feature_test , target_train , target_test = train_test_split(features , targets , train_size = 0.2 , random_state=0)

model = Sequential()
model.add(Dense(10 , input_dim = 3 , activation = 'sigmoid'))
model.add(Dense(2 , activation = 'softmax'))

optimizer = Adam(learning_rate = 0.001)
model.compile(loss = 'categorical_crossentropy' , metrics = ['accuracy'] , optimizer = optimizer)

model.fit(feature_train , target_train , epochs = 1000 , verbose = 2)
```

```python
results = model.evaluate(feature_test , target_test , use_multiprocessing = True)
print(results)
```
